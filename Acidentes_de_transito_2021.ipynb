---
**Introdução:**

Acidentes de trânsito representam importante problema de saúde pública global, demandando esforços conjuntos para a prevenção eficaz e sustentável.

***Infelizmente, o Brasil ocupa a quinta posição entre os países com maior número de mortes por acidentes de trânsito.***

---

**Objetivo:**

Analisar os acidentes nas rodovias federais do Brasil no ano de 2021 e responder às seguintes perguntas:

1- Estados Brasileiros onde ocorreram mais acidentes de trânsito nas rodovias em 2021

2- Rodovias Brasileiras onde ocorreram mais acidentes de trânsito em 2021

3- Principais causas de acidentes de trânsito nas rodovias Brasileiras em 2021

4- Dias da semana que ocorreram mais acidentes de trânsito nas rodovias do Brasil em 2021

5- Fases do dia que ocorreram mais acidentes de trânsito nas rodovias do Brasil em 2021

6- Tipos de acidentes de trânsito que mais ocorreram nas rodovias do Brasil em 2021

7- Condições metereológicas que mais ocorreram acidentes de trânsito nas rodovias do Brasil em 2021

8- Tipos de veículos que mais se envolveram em acidentes de trânsito nas rodovias do Brasil em 2021

9- Identidade sexual das pessoas com maior envolvimento em acidentes de trânsito nas rodovias do Brasil em 2021

10- Gravidade dos acidentes de trânsito que mais ocorreram nas rodovias do Brasil em 2021

---

```python
#Importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Abrir arquivo (CSV) e montar o data frame (df)
df = pd.read_csv('prf_acidentes_2021.csv', delimiter=',', encoding='utf-8')
```

```python
#Exibe quantidades de linhas e colunas do data frame
df.shape
```

```python
#Exibir as lindas com valores faltentes
df.isnull().sum()
```

```python
#Exibir as (...) primeiras linhas. O padrão "()", exibe 5. 
df.head()
```

```python
#Gráfico: Estados Brasileiros onde ocorreram mais acidentes de trânsito nas rodovias em 2021
plt.figure(figsize=(20,10))
plt.title('Estados Brasileiros onde ocorreram mais acidentes de trânsito nas estradas em 2021.', fontweight='bold')
plt.xlabel('Qtde. acidentes', fontweight='bold')
plt.ylabel('Estados', fontweight='bold')
fig = df['uf'].value_counts(ascending=True).plot.barh();
plt.show(fig)
```

```python
#Gráfico: Rodovias Brasileiras onde ocorreram mais acidentes de trânsito em 2021
plt.figure(figsize=(20,10))
plt.title('Rodovias Brasileiras onde ocorreram mais acidentes de trânsito em 2021.', fontweight='bold')
plt.xlabel('Rodovias', fontweight='bold')
plt.ylabel('Qtde. acidentes', fontweight='bold')
fig = df['br'].value_counts(ascending=False).plot.bar();
plt.show(fig)
```

```python
#Gráfico: Principais causas de acidentes de trânsito nas rodovias Brasileiras em 2021
plt.figure(figsize=(20,5))
plt.title('Principais causas de acidentes de trânsito nas rodovias Brasileiras em 2021.', fontweight='bold')
plt.xlabel('Principais causas de acidentes', fontweight='bold')
plt.ylabel('Qtde. acidentes', fontweight='bold')
fig = df['causa_acidente'].value_counts(ascending=False).plot.bar();
plt.show(fig)
```

```python
#Gráfico: Dias da semana que ocorreram mais acidentes de trânsito nas rodovias do Brasil
plt.figure(figsize=(20,10))
plt.title('Dias da semana que ocorreram mais acidentes de trânsito nas rodovias do Brasil em 2021.', fontweight='bold')
plt.xlabel('Dias da semana', fontweight='bold')
plt.ylabel('Qtde. acidentes', fontweight='bold')
fig = df['dia_semana'].value_counts(ascending=False).plot.bar();
plt.show(fig)
```

```python
#Gráfico: Fases do dia que ocorreram mais acidentes de trânsito nas rodovias do Brasil
plt.figure(figsize=(20,10))
plt.title('Fases do dia que ocorreram mais acidentes de trânsito nas rodovias do Brasil em 2021.', fontweight='bold')
#plt.xlabel('Dias da semana', fontweight='bold')
#plt.ylabel('Fase do dia',fontweight='bold')
fig = df['fase_dia'].value_counts(ascending=False).plot.pie();
plt.show(fig)
```

```python
#Gráfico: Tipos de acidentes de trânsito que mais ocorreram nas rodovias do Brasil
plt.figure(figsize=(20,10))
plt.title('Tipos de acidentes de trânsito que mais ocorreram nas rodovias do Brasil em 2021.', fontweight='bold')
plt.xlabel('Qtde.', fontweight='bold')
plt.ylabel('Tipos de acidentes',fontweight='bold')
fig = df['tipo_acidente'].value_counts(ascending=True).plot.barh();
plt.show(fig)
```

```python
#Gráfico: Condições metereológicas que mais ocorreram acidentes de trânsito nas rodovias do Brasil
plt.figure(figsize=(20,10))
plt.title('Condições metereológicas que mais ocorreram acidentes de trânsito nas rodovias do Brasil em 2021.', fontweight='bold')
plt.xlabel('Condições metereológicas', fontweight='bold')
plt.ylabel('Qtde.',fontweight='bold')
fig = df['condicao_metereologica'].value_counts(ascending=False).plot.bar();
plt.show(fig)
```

```python
#Gráfico: Tipos de veículos que mais se envolveram em acidentes de trânsito nas rodovias do Brasil em 2021
plt.figure(figsize=(20,10))
plt.title('Tipos de veículos que mais se envolveram em acidentes de trânsito nas rodovias do Brasil em 2021.', fontweight='bold')
plt.xlabel('Qtde.', fontweight='bold')
plt.ylabel('Tipos de veículos',fontweight='bold')
fig = df['tipo_veiculo'].value_counts(ascending=True).plot.barh();
plt.show(fig)
```

```python
#Gráfico: Identidade sexual das pessoas com maior envolvimento em acidentes de trânsito nas rodovias do Brasil em 2021
plt.figure(figsize=(20,10))
plt.title('Identidade sexual das pessoas com maior envolvimento em acidentes de trânsito nas rodovias do Brasil em 2021.', fontweight='bold')
#plt.xlabel('Qtde.', fontweight='bold')
#plt.ylabel('Tipos de veículos',fontweight='bold')
fig = df['sexo'].value_counts(ascending=True).plot.pie();
plt.show(fig)
```

```python
#gráfico: Gravidade dos acidentes de trânsito que mais ocorreram nas rodovias do Brasil em 2021
plt.figure(figsize=(20,10))
plt.title('Gravidade dos acidentes de trânsito que mais ocorreram nas rodovias do Brasil em 2021.', fontweight='bold')
plt.xlabel('Gravidade dos acidentes', fontweight='bold')
plt.ylabel('Qtde.',fontweight='bold')
fig = df['estado_fisico'].value_counts(ascending=False).plot.bar();
plt.show(fig)
```
